```python
a1=np.array([11,16,4,3,2,8,0,6])
a3=a1.reshape(-1,1)
a3 # -1= no of rows free, but ensure that no of columns is 1
a2=a1.reshape(-1,2) # no of rows free(python will decide), but 2 columns
a2
#array([[11, 16],
#       [ 4,  3],
#       [ 2,  8],
#       [ 0,  6]])

# print row 2
a2[1] # index starts with 0 , so 0 1
a2[1,:] #
a2[:,1] # 2nd column


a2[[2,3],:] #multiple rows  , will give rows with index 2 and 3

a2[:,[0,1]] # multiple columns , column with index 0 and 1
```

```python
newcol=np.array([3,1,7,9])
newcol=newcol.reshape(-1,1)
newcol # row of values converted in column values
```

```python
# hstack -- adding column , array will grow horizontally so hstack
# vstack -- adding row , array will grow vertically so vstack
```

```python
a3=np.hstack([a2,newcol]) # added new column  --- hstack([array,colname])
a3

newrow=np.array([2,3,4])
a3=np.vstack([a3,newrow]) # added new row --vstack([array,rowname])
a3

a3=np.insert(a3,3,[23,56,77],axis=0) # adding row at specific position, [arrayname,index to insert,values,axis=0 tells row]
a3

a3=np.insert(a3,2,[11,12,13,14,15,16],axis=1)
a3

# delete/ remove row index 1,3
a4=np.delete(a3,(1,3),axis=0)
a4
#delete / remove col index 1,3
a5=np.delete(a3,(1,3),axis=1)
a5
```

```python
b=np.array([2,4,1,7,6,5,9,8,2,3,1,5])
b
c = sorted(b)
c      #sorts the array in increasing order
c = sorted(b)[::-1]    # sorts the array in the decreasing order
c
np.sort(b)     # also PRINTS the array in the increasing order, original array is unchanged

b=b.reshape(3,4)   # reshapes b as a 2d array with 3 rows and 4 column
np.sort(b,axis=0)   # row wise values will be changed. Column values are only rearranged
np.sort(b,axis=1)   # column wise values will be changed. Row values are only rearranged

# axis=0  - top row will have least value
            #bottom row will have all highest values

# axis=1  - leftmost column will have least values
           #rightmost columns will have highest values
```
