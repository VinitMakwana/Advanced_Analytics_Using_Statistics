```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('../data/CDAC_DataBook.xlsx', sheet_name='ERPData')
df.head()
```

```python
# Extracting one columns from dataset
x1 = df.MaterialID
x1.head()
```

```python
# length of unique materialid
len(np.unique(x1))
```

```python
np.unique(x1, return_counts=True)
```

```python
# Counting the number of unique material ids
x1.value_counts()
```

```python
# Converting label (materialid) into a column using reset_index()
df1 = x1.value_counts().reset_index()
df1
```

```python
# Proportion of each material id
df1['prop'] = df1['count']/sum(df1['count'])
df1
```

```python
# Percentate proportion of each material id
df1['pct'] = df1['prop']*100
df1
```

```python
# Cumulative percentage for each material id
df1['Cumpct'] = np.cumsum(df1['pct'])
df1
```

```python
# Sorting the material id based on count
df1.sort_values('count',ascending=False)
```
